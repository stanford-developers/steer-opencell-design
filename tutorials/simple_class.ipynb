# A simple class with all the key ingredients

```python
import numpy as np

M_TO_CM = 1e2
CM_TO_M = 1e-2


class _BaseClass:

    def __init__(
            self,
            a: float,
            b: float, 
        ) -> None:
        """
        A Base class that serves as an example for inheritance.
        """
        self.a = a
        self.b = b

    def _calculate_area(self) -> float:
        """
        A private method to calculate the area.
        """
        self._area = self._a * self._b

    @property
    def a(self) -> float:
        return round(self._a * M_TO_CM, 5)
    
    @a.setter
    def a(self, value: float) -> None:
        
        self._a = value * CM_TO_M

        if hasattr(self, '_b'):
            self._calculate_area()

    @property
    def b(self) -> float:
        return round(self._b * M_TO_CM, 5)
    
    @b.setter
    def b(self, value: float) -> None:
        
        self._b = value * CM_TO_M
        
        if hasattr(self, '_a'):
            self._calculate_area()
    
    @property
    def area(self) -> float:
        return round(self._area * M_TO_CM**2, 5)

    @area.setter
    def area(self, value: float) -> None:
        
        scale_factor = np.sqrt(value / self.area)

        self.a = scale_factor * self.a
        self.b = scale_factor * self.b
        self._area = value * CM_TO_M**2
    
    
class Rectangle(_BaseClass):
    """
    A class that inherits from _BaseClass and adds some functionality.
    """
    
    def __init__(
            self, 
            a: float, 
            b: float
        ) -> None:
        
        super().__init__(a, b)
    

```

```python

my_instance = Rectangle(a = 10, b = 20)
```

```python
print(f"a: {my_instance.a} cm")
print(f"b: {my_instance.b} cm")
print(f"area: {my_instance.area} cm^2")
```

```python
my_instance.area = 100
print(f"a: {my_instance.a} cm")
print(f"b: {my_instance.b} cm")
print(f"area: {my_instance.area} cm")
```
