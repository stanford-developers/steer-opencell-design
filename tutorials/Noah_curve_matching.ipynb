```python
from OpenCell.Materials.RawMaterials import CurrentCollectorMaterial
from OpenCell.Materials.CurrentCollectors import TablessCurrentCollector
from OpenCell import __version__

import pandas as pd
from datetime import datetime as dt
```

```python

cathode_current_collector_material = CurrentCollectorMaterial.from_database('Aluminum')

cathode_current_collector = TablessCurrentCollector(
    material=cathode_current_collector_material,
    width=151,
    length=2975,
    coated_width=142,
    insulation_width=3,
    thickness=13,
    name='QNAS-S40160NL'
)

```

```python

from OpenCell.Materials.ElectrodeMaterials import CathodeMaterial, ConductiveAdditive, Binder
from OpenCell.Formulations.ElectrodeFormulations import CathodeFormulation, AnodeFormulation

cathode_active_material1 = CathodeMaterial.from_database("NaNiMn P2-O3 Composite")
cathode_active_material1.irreversible_capacity_scaling = 1
cathode_active_material1.reversible_capacity_scaling = 1


cathode_conductive_additive1 = ConductiveAdditive.from_database("Super P")
cathode_conductive_additive2 = ConductiveAdditive.from_database("Graphite")

cathode_binder1 = Binder.from_database("PVDF")
cathode_binder2 = Binder.from_database("CMC")


cathode_formulation = CathodeFormulation(
    active_materials={cathode_active_material1: 90},
    binders={cathode_binder1: 3, cathode_binder2: 2},
    conductive_additives={cathode_conductive_additive1: 3, cathode_conductive_additive2: 2}
)



from OpenCell.Constructions.Electrodes import Cathode
from OpenCell.Materials.RawMaterials import InsulationMaterial


insulation_material = InsulationMaterial.from_database("Aluminium Oxide, 95%")

my_cathode = Cathode(
    formulation=cathode_formulation,
    current_collector=cathode_current_collector,
    calender_density=3,
    mass_loading=15,
    insulation_material=insulation_material,
    insulation_thickness=3
)

my_cathode.half_cell_curve
```
