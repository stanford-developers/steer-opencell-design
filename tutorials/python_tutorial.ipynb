# Material for our current collector

### Creating cathode current collector material

```python
cathode_cc_material = CurrentCollectorMaterial(
    name='Aluminum',
    density=2.7,
    specific_cost= 0.01,
    color='silver',
)
```

### or you can get it form a database

```python
cathode_current_collector_material = CurrentCollectorMaterial.from_database(name='Aluminum')
```

### We can take a look at the properties and attributes of the material

```python
cathode_current_collector_material
```

### We can print out all its useful attributes

```python
print(f'Cathode cc material density: {cathode_current_collector_material.density} g/cm^3')
print(f'Cathode cc material specific cost: {cathode_current_collector_material.specific_cost} $/kg')
print(f'Cathode cc material color: {cathode_current_collector_material.color}')
```

### These properties call on the material's getter methods, which are defined in the `_RawMaterial` class. 

The getter method finds the internally stored value and returns it in reasonable units. Each property has a mirror value stored internally and privately, recorded in SI units.

```python
print(f'Cathode cc material density: {cathode_current_collector_material._density} kg/m^3')
print(f'Cathode cc material specific cost: {cathode_current_collector_material._specific_cost} $/kg')
print(f'Cathode cc material color: {cathode_current_collector_material._color}')
```

### We can set any of these attributes. 

The internal logic is handled by that attribute's setter method. This allows all internal attribute to be updated corrospondingly. 

```python
cathode_current_collector_material.density = 3.1

print(f'Cathode cc material density: {cathode_current_collector_material.density} kg/m^3')
```

### Let's make a current collector with this material

```python

cathode_current_collector = PunchedCurrentCollector(
    name='Aluminum Punched Current Collector',
    material=cathode_current_collector_material,
    thickness=12,
    width=180,
    height=200,
    tab_width=20,
    tab_height=20,
    tab_position=30,
    coated_tab_height=4,
    insulation_width=6
)
```

# Electrode Formulation

### Lets grab some cathode materials, binders and conductive additives from the database. Remember, we could make these in a similar way as the current collector material.  In the case of the active materials however, we would need to include their electrochemical curves.

```python
cathode_material_1 = CathodeMaterial.from_database(name='NMC811')
cathode_material_1.extrapolation_window = 0.3

cathode_material_2 = CathodeMaterial.from_database(name='LFP')

my_binder = Binder.from_database(name='PVDF')
my_conductive_additive = ConductiveAdditive.from_database(name='Super P')
```

### Lets make a formulation with these materials

```python

my_formulation = CathodeFormulation(
    active_materials={
        cathode_material_1: 55,
        cathode_material_2: 40
    },
    binders={
        my_binder: 3
    },
    conductive_additives={
        my_conductive_additive: 2
    }
)
```

### Lets look at the formulations properties

```python
my_formulation
```

```python

cathode = Cathode(
    formulation=my_formulation,
    current_collector=cathode_current_collector,
    mass_loading=10.2,
    calender_density=1.5,
    insulation_material=InsulationMaterial.from_database(name='Aluminium Oxide, 95%'),
    insulation_thickness=10
)
```
